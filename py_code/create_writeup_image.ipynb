### undistortion

```python
from find_lane import FindLane
import matplotlib.pyplot as plt 
import numpy as np 
import cv2
%matplotlib inline 

ym_per_pix = 30/720
xm_per_pix = 3.7/700
find_lane = FindLane(ym_per_pix=ym_per_pix, xm_per_pix=xm_per_pix)
find_lane.calcDistort()

img = cv2.imread('../camera_cal/calibration1.jpg')
undist_img = find_lane.undistImage(img)

plt.figure(figsize=(12, 8))
plt.subplot(121)
plt.title('Original image')
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.subplot(122)
plt.title('Undistortion image')
plt.imshow(cv2.cvtColor(undist_img, cv2.COLOR_BGR2RGB))
# plt.savefig('../writeup_images/undist_chessboard.jpg', dpi=70)


```

### undistort test image

```python
img = cv2.imread('../test_images/test1.jpg')
undist_img = find_lane.undistImage(img)
# cv2.imwrite('../writeup_images/undist_test1.jpg', undist_img)
plt.imshow(cv2.cvtColor(undist_img, cv2.COLOR_BGR2RGB))
plt.show()
```

### apply color and gradient threshold

```python
find_lane.createColorGrad('BGR')
combined_color, combined_binary = find_lane.color_grad.colorGradPipeline(undist_img, auto_finding_saturation_thresh=True)

combined_color = cv2.cvtColor(combined_color.astype(np.uint8), cv2.COLOR_RGB2BGR)
# cv2.imwrite('../writeup_images/color_gradient_threshold.jpg', combined_color)
plt.imshow(combined_binary, cmap='gray')
plt.show()
```

### perspective transform

```python
src = np.float32([[685, 450], [1120, 720], [190, 720], [595, 450]])
dst = np.float32([[900, 0], [900, 720], [320, 720], [320, 0]])
warped_size=(int(img.shape[1]), int(img.shape[0]))
vertices = np.array([[(0,800) ,(0, 0), (1200, 0), (1200, 800)]], dtype=np.int32)
find_lane.createWarp(src, dst, warped_size, vertices)
drawn_region = find_lane.warp.drawRegion(undist_img)
birds_img = find_lane.warp.birdsEye(drawn_region)

plt.figure(figsize=(12, 5))
plt.subplot(121)
plt.title('Undistorted image with source points drawn')
plt.imshow(cv2.cvtColor(drawn_region, cv2.COLOR_BGR2RGB))
plt.subplot(122)
plt.title('Warped result with destinaion points drawn')
plt.imshow(cv2.cvtColor(birds_img, cv2.COLOR_BGR2RGB))
# plt.savefig('../writeup_images/perspective_trans.jpg', dpi=80)

```

### find lane pixels with histogram

```python
binary_warped = find_lane.warp.birdsEye(combined_binary)
plt.figure(figsize=(5, 8))
plt.subplot(211)
plt.title('Warped binary image')
plt.imshow(binary_warped, cmap='gray')
plt.subplot(212)
plt.title('column direction sum up')
histogram = np.sum(binary_warped[binary_warped.shape[0]//2:,:], axis=0)
plt.plot(histogram)
center = binary_warped.shape[1]//2
plt.plot([center, center], [0, 500], label='center of image')
plt.legend()
# plt.savefig('../writeup_images/binary_histogram.jpg', dpi=80)
plt.show()
```

### sliding window

```python
import matplotlib.image as mpimg
binary_warped = binary_warped * 255
leftx, lefty, rightx, righty, out_img = find_lane.find_lane_pixels(binary_warped)
# mpimg.imsave('../writeup_images/sliding_window.jpg', out_img.astype(np.uint8))
plt.imshow(out_img)
plt.show()
```

### fit polynomial

```python
poly_img, curvature, ideal_car_pos = find_lane.fit_polynomial(binary_warped)
# mpimg.imsave('../writeup_images/fit_polynomial.jpg', poly_img.astype(np.uint8))
plt.imshow(poly_img)
plt.show()
```
